```python
import pandas as pd
import numpy as np
```

```python
deliveries=pd.read_csv(r"D:\ML\Projects\IPL\deliveries.csv")
matches=pd.read_csv(r"D:\ML\Projects\IPL\matches.csv")
```

```python
matches.head()
```

```python
deliveries.head()
```

```python
deliveries_df=deliveries[['match_id','inning','batting_team','bowling_team','over','ball','total_runs','is_wicket']]
deliveries_df=deliveries_df[deliveries_df['inning']==2]
deliveries_df
```

```python
matches_df=matches[['id','city','winner','target_runs','target_overs','super_over']]
matches_df=matches_df[matches['super_over']=='N']
matches_df
```

```python
matches_df['winner'].unique()
```

```python
teams=[
    'Sunrisers Hyderabad',
    'Mumbai Indians',
    'Royal Challengers Bangalore',
    'Kolkata Knight Riders',
    'Punjab Kings',
    'Chennai Super Kings',
    'Rajasthan Royals',
    'Delhi Capitals',
    'Gujarat Titans',
    'Lucknow Super Giants'
]
```

```python
matches_df['winner']=matches_df['winner'].str.replace('Delhi Daredevils','Delhi Capitals')
matches_df['winner']=matches_df['winner'].str.replace('Kings XI Punjab','Punjab Kings')
matches_df['winner']=matches_df['winner'].str.replace('Deccan Chargers','Sunrisers Hyderabad')

deliveries_df['batting_team']=deliveries_df['batting_team'].str.replace('Delhi Daredevils','Delhi Capitals')
deliveries_df['batting_team']=deliveries_df['batting_team'].str.replace('Kings XI Punjab','Punjab Kings')
deliveries_df['batting_team']=deliveries_df['batting_team'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
deliveries_df['bowling_team']=deliveries_df['bowling_team'].str.replace('Delhi Daredevils','Delhi Capitals')
deliveries_df['bowling_team']=deliveries_df['bowling_team'].str.replace('Kings XI Punjab','Punjab Kings')
deliveries_df['bowling_team']=deliveries_df['bowling_team'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
```

```python
matches_df=matches_df[matches_df['winner'].isin(teams)]
matches_df.shape
```

```python
deliveries_df=deliveries_df[deliveries_df['batting_team'].isin(teams)]
deliveries_df=deliveries_df[deliveries_df['bowling_team'].isin(teams)]
deliveries_df.shape
```

```python
matches_df=matches_df[matches_df['target_overs']==20]
matches_df
```

```python
deliveries_df
```

```python
data=pd.merge(matches_df[['id','city','winner','target_runs']],deliveries_df,left_on='id',right_on='match_id')
data.head(10)
```

```python
data.batting_team.unique()
```

```python
data['curr_score']=data.groupby('match_id')['total_runs'].cumsum()
data['runs_left']=data['target_runs']-data['curr_score']
data['balls_left']=120-(data['over']*6+data['ball'])
```

```python
data
```

```python
wickets=data.groupby('match_id')['is_wicket'].cumsum()
data['wickets_left'] =10-wickets
data
```

```python
data['nrr']=(data['curr_score']*6)/(data['over']*6+data['ball'])
data['rrr']=(data['runs_left']*6)/(data['balls_left'])
data
```

```python
def result(obj):
    if obj.batting_team==obj.winner:
        return 1
    return 0
data['result']=data.apply(result,axis=1)
```

```python
df=data[['city','batting_team','bowling_team','target_runs','runs_left','balls_left','wickets_left','nrr','rrr','result']]
df=df[df['balls_left']!=0]
df
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
df
```

```python
df=df.sample(df.shape[0])
df
```

```python
df.city.unique()
```

```python
X=df.drop('result',axis='columns')
Y=df.result
```

```python
from sklearn.model_selection import train_test_split
from sklearn import linear_model
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.2)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline

trf=ColumnTransformer([
    ('trf',OneHotEncoder(drop='first'),['batting_team','bowling_team','city'])
]
,remainder='passthrough')

pipe=Pipeline(steps=[
    ('step1',trf),
    ('step2',LogisticRegression(solver='liblinear'))
])
```

```python
X_test.columns
```

```python
pipe.fit(X_train,y_train)
```

```python
y_prediction=pipe.predict(X_test)
```

```python
pipe.score(X_test,y_test)
```

```python
pipe.predict_proba(X_test)[2]
```
